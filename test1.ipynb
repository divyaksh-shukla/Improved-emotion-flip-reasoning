# EFR

```python
with open('data/MELD_val_efr.json', 'rb') as f:
    file_data = f.read().decode('cp1252')
```

```python
with open('data/MaSaC_train_efr.json', mode='rb') as f:
    file_data = f.read().decode('cp1252')
    data = json.loads(file_data)
    data = pd.DataFrame(data)

# data = pd.read_json('data/MaSaC_train_efr.json', encoding='latin1')
data = data.reset_index(names=['flip_id'])
data.head(10)
```

```python
def dialog_converter(data, index):
    json_data = data.iloc[index][['flip_id', 'speakers', 'utterances', 'emotions', 'triggers']].to_json()
    json_data = json.loads(json_data)
    df = pd.DataFrame(json_data)
    df = df.replace({'triggers': {'': '0.0', 'Curiosity': '0.0', 'Annoyance': '0.0'}})
    return df.astype({'triggers': 'float64'})

all_dialogs = pd.concat([dialog_converter(data, i) for i in range(len(data))])
all_dialogs
```

```python
all_dialogs['dialog_id'] = 0
def imbed_dialog_ids(data):
    prev_first_utt = data.loc[data['flip_id'] == 0, 'utterances'][0]
    dialog_id = 0
    dialog_lengths = np.zeros(452)
    for fid in range(1, 4893):
        try:
            if data.loc[data['flip_id'] == fid, 'utterances'][0] != prev_first_utt:
                prev_first_utt = data.loc[data['flip_id'] == fid, 'utterances'][0]
                dialog_id += 1
            data.loc[data['flip_id'] == fid, 'dialog_id'] = dialog_id
            dialog_lengths[dialog_id] = len(data.loc[data['flip_id'] == fid, 'utterances'])
        except Exception as e:
            print(fid)
            raise e
    return data, dialog_lengths
all_dialogs, dialog_lengths = imbed_dialog_ids(all_dialogs.copy())
all_dialogs
```

```python
def convert_utterances_to_cp1252(data):
    data = data.reset_index(drop=True)
    try:
        for i in range(len(data)):
            data.loc[i, 'utterances'] = data.loc[i, 'utterances'].encode('utf-8').decode('cp1252')
    except Exception as e:
        print(data.loc[i, 'utterances'], i)
        raise e
    return data

all_dialogs = convert_utterances_to_cp1252(all_dialogs)
all_dialogs
```
